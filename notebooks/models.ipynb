#### Customizing Train and Test Sets

25 experiments were relabeled by Pavel. Here I use these labels to update labels loaded above.

Note that if the labels on train set are of an improved quality, then the ones on test set will be of poorer quality. This means that you should not expect stellar performance on test.

**RUN THIS ONLY IF YOU WANT TO CHANGE THE SETS LOADED BEFORE.**

````python
import numpy as np
import pandas as pd
import random

 # how many additional experiments should be added to training set?
ENRICH_NEXP = 50

# read your updated train labels
flabels_new = "/home/mholub/tmp/mlQC/labels/to_verify/labels_corrections2.csv"
y_train = []
exps = set()
label_map = lambda x: 1 if x == "good" else 0

with open(flabels_new, "r") as f:
    for row in f:
        line = list(filter(bool, row.strip().split(","))) # strip whitespace, split, drop empty
        sample = "_".join(line[0:2]) # combine exp and sample name
        exps.add(line[0]) # store experiments
        line[-1] = label_map(line[-1])
        line[-2] = int(line[-2])
        y_train.append([sample] + line[2:])

y_train = np.array(y_train)


# combine previously disjoint, this is for convenience and compatibility
# could also use just one of the two.
X_all = np.concatenate((X, X_test), axis = 0) # (X, X_test)
y_all = pd.concat((y, y_test), axis = 0) # (y, y_test)

# Enrich training set with other random examples
all_exps = set(map(lambda x: x.split("_")[0], y_all.index))
exps.update(set(random.sample(all_exps, ENRICH_NEXP)))

# Make spliting index
exps_tup = tuple(exps)
drop_id = np.asarray([False if x.startswith(exps_tup) else True for x in y_all.index])

# Split dataset into  two
X1 = X_all[~drop_id, :]
X_test1 = X_all[drop_id, :]

# Make sure you pass in int, remap values to new ones 
y1_checked = pd.Series(data = list(map(int, y_train[:, -1])), index = y_train[:, 0])
y1 = y_all[~drop_id]
y_test1 = y_all[drop_id]

# Update or create new values, depending on how you construct y_all
y1.update(y1_checked)
# y1[y1.index.isin(y_train[:, 0])] = list(map(int, y_train[:, -1])) # update values

# lets, say that you actually want to preserve just one type of data for the test data you loaded
# (either fully or partially bad QC; change this by inverting drop_id)

do_drop_set = False
if do_drop_set:
    drop_samples = set(samples) # relies on order from above
    drop_id = np.asarray([False if x in drop_samples else True for x in y_test1.index])
    y_test1 = y_test1[drop_id]
    X_test1 = X_test1[drop_id, :]
else:
    assert X1.shape[0] + X_test1.shape[0] == X_all.shape[0]

# Assert disjoint and complete
assert set.isdisjoint(set(y_test1.index), set(y1.index))


# reassing after you have played with labels
try:
    X, X_test = X1, X_test1
    y, y_test = y1, y_test1
except:
    pass

del X1, y1, X_test1, y_test1
````

### Train Test Predict

````python
# On deoployment you will be feeding in data that has full set of features
# and may not have the information which features were used for training the model.
# Here you enable the pipeline to deal with this.
# Note that feature vectors must have same length on training and test.
try:
    kept_features = np.argwhere(~reader.data_reader.feature_drop_id)
except AttributeError as e:
    kept_features = np.argwhere(~reader.feature_drop_id)

# Pipeline is a chain of feature selectors,transformers and estimators. Here you control
# what is in the chain and some parameters thereof. Absolute majority of parameters 
# are selected automatically.
pipe_cfg = {
    "feature_transformation": {
        "do_threshold": True, # threshold out constant features
        "do_scale": True , # StandardScaler, should be always ON
        "params": {"val_thresh": None, # allow any magnitude of feature
                  "kept_features_id": kept_features}}, # use this to subset if longer vec
    "feature_selection": {"which": [0], # 0 is none (prefered); FPR, FWE worked
                          },
    # 1 and/or 2 (possibly combined with 5) work.
    # Prefer 5 ON to do probability calibraiton
    "classification": {
        "which": [2, 5],
        "weights": "random", # Will search over weights, prefered.
    },
    "n_iter": 30,
}

# Construct model (Pipeline wrapped in RandomizedSearchCV)
model_ = model.Model(y, random_state = 56, verbose = 0)
search_cv_pipe = model_.make_search_cv(pipe_cfg)

mt = trainer.ModelTrainer( model = search_cv_pipe, verbose= model_.verbose,
                          rnd = model_.random_state, allow_calibration = False,
                         save_folder = "/home/mholub/tmp/mlQC/saved/saved", 
                         refit = False)

# Train and test
test_score = mt.train_test([X, X_test], [y, y_test])

# Get predicitons (with probability) used for test
y_hat = mt.predict(X_test, proba = True)
# Get wrong, correct predicitons as DatFrame ordered by increasing certainity 
wrong, correct = mt.get_prediction_certainity()
````

````python
# Plot leading dimension driving decision in pair plot
# This will highlight which features are deemed important to the model.
# Use this together with the vis.plot_features to figure out which features to feed.
# Use this to interpret your model.

from maqc.visualization.vis import Visualizer
from functools import reduce
import numpy as np

vis = Visualizer()

plot_raw = False

support = mt.get_support() # Which features were retained in feature selection?
importance = mt.feature_importances_ # Importance on classification

try:
    reader_ = reader.data_reader
except AttributeError as e:
    reader_ = reader

if not plot_raw:
    try:
        est = mt.estimator.base_estimator
    except AttributeError as e:
        est = mt.estimator

    # Obtain chained feature transformation/selection
    funs = []
    for name, st in est.steps:
        if name not in {"scaler", "feature_selection", "val_thresholder", "var_thresholder"}:
            continue
        obj = getattr(est.named_steps, name)
        funs.append(obj.transform)
    # Construct function applying chained feature transformation/selection
    transform_ = lambda x: reduce(lambda res, f: f(res), funs, x)
    
    # Apply chained feature transformation/selection
    XX = [  transform_(X), transform_(X_test)]

    if np.std(importance) == 0:
        which = "PCA" # ICA, random, ...
        print("Features do not have assigned importances, plotting {} projection.".format(which))
        feat_names = None
    else:
        which = np.argsort(importance)[::-1][:4]
        df, (feat_names, _) = reader_.interpret_features(support[:], importance)

else:
    # Plot raw data
    XX = [X, X_test]
    df, (_, feat_names) = reader_.interpret_features(support[:], importance,
                                                               use_untransformed_index = True)
    which = df.index.tolist()[:4]
        

vis.plot_leading_dims(XX, [y, y_test], dim_names = feat_names, which = which)
````

````python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA, KernelPCA
from sklearn.preprocessing import StandardScaler
from sklearn.random_projection import GaussianRandomProjection as GRP
import seaborn as sns
import pandas as pd
import numpy as np

def dim_reduce(X, ncomp = 4, what = "pca", scale = True):
    """Seems not to be doing harm and speeds up"""
    if scale: X = StandardScaler().fit_transform(X)
    
    if what == "pca":
        pca = PCA(n_components = ncomp, whiten= False)
        Xout = pca.fit_transform(X)
        print("Explained variance ratio: {}".format(pca.explained_variance_ratio_))
    
    elif what == "ica":
        ica = FastICA(n_components = ncomp, max_iter = 10000)
        Xout = ica.fit_transform(X)
    
    elif what == "random":
        grp = GRP(n_components = ncomp)
        Xout = grp.fit_transform(X)
    
    elif what == "kpca":
        kpca = KernelPCA(n_components = ncomp, kernel = "rbf", max_iter = 10000)
        Xout = kpca.fit_transform(X)
        
    return (Xout)
````

### Sample Size Wrapper (NOT UPDATED)

It looks like the model does not scale well with increasing number of samples. Below is simple evidence of thereof. It is advisable to repeat the experiment couple of times to collect more score values for each value of number of samples.

````python
%load_ext autoreload
%autoreload 2
from maqc.io import readers as io
from maqc.models import trainer, model
````

````python
def make_trainer(y):
    pipe_cfg = {
    # passing list of indices combines estimators 
    "scaler": True,
    "feature_selection": [4, 6], # FPR, FWE work very well
    "classification": [2],
    "weights": "random",
    }
    
    model_ = model.Model(y, random_state = 56, verbose = 2)
    search_cv_pipe = model_.make_search_cv(pipe_cfg)

    mt = trainer.ModelTrainer( model = search_cv_pipe, verbose= model_.verbose,
                              rnd = model_.random_state, allow_calibration = True)
    return mt

def load_train_test(n):
    dir_path = "/home/mholub/tmp/mlQC/data/data"
    dr = io.DataReader(load_dir = dir_path)
    samples, X = dr.load_sample_data(nsamples = n)

    flabels = "/home/mholub/tmp/mlQC/labels/paritaly_bad_exps_samples_final.tsv"
    y, keepid = io.LabelReader(samples, flabels = flabels).read_labels(include_exp = True)
    X = X[keepid, :]
    
    mt = make_trainer(y)
    
    test_score, _,_ = mt.train_test(X, y)
    return test_score
````

````python
test_scores = []
nsamples = [300, 500, 1000, 2000, 5000]
num_trials = 5
for ns in nsamples:
    ts = []
    for _ in range(num_trials):
        ts.append(load_train_test(ns))
    test_scores.append(ts)
````
